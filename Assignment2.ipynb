3. Debug common CUDA errors (zero output, incorrect indexing, PTX errors)



Zero Output Error:
Occurs when the CUDA kernel does not execute properly, device synchronization is missing, or results are not copied back from device to host memory.
It can be debugged by verifying kernel launch parameters, using cudaDeviceSynchronize()

Incorrect Indexing Error:
Happens when thread indexing is calculated incorrectly, leading to wrong or repeated outputs.
It can be resolved by using correct indexing formulas and boundary checks.

PTX Errors:
PTX - Parallel Thread Execution errors occur due to CUDA version mismatch, unsupported GPU architecture, or invalid compiler options.
These errors can be debugged by matching CUDA toolkit and driver versions and compiling for the correct GPU architecture.



4. Write a CUDA C/C++ program to demonstrate GPU kernel execu'on and thread indexing.
a. Launch a CUDA kernel using: 1 block and 8 threads
b. Each thread must print: Hello from GPU thread <global_thread_id>
c. Compute the global thread ID using: global_thread_id = blockIdx.x * blockDim.x +
threadIdx.x
d. Clearly separate: Host code (CPU) & Device code (GPU kernel)

```python
%%writefile test.cu
#include <stdio.h>
#include <cuda_runtime.h>

/* -------- Device Code (GPU Kernel) -------- */
__global__ void helloKernel() {
    int global_thread_id = blockIdx.x * blockDim.x + threadIdx.x;
    printf("Hello from GPU thread %d\n", global_thread_id);
}

/* -------- Host Code (CPU) -------- */
int main() {

    helloKernel<<<1, 8>>>();

    // Check kernel launch error
    cudaError_t err = cudaGetLastError();
    if (err != cudaSuccess) {
        printf("CUDA error: %s\n", cudaGetErrorString(err));
    }

    cudaDeviceSynchronize();
    cudaDeviceReset();

    return 0;
}
```

```python
!nvcc -arch=sm_75 test.cu -o test
```

Write a CUDA program to demonstrate host and device memory separa'on.
a. Create an integer array of size 5 on the host (CPU).
b. Allocate corresponding memory on the device (GPU) using cudaMalloc().
c. Copy data from host to device using cudaMemcpy().
d. Launch a kernel where GPU threads print values from device m

```python
%%writefile memory_sep.cu
#include <stdio.h>
#include <cuda_runtime.h>

/* -------- Device Code (GPU Kernel) -------- */
__global__ void printDeviceArray(int *d_arr) {
    int idx = threadIdx.x;
    if (idx < 5) {
        printf("GPU: d_arr[%d] = %d\n", idx, d_arr[idx]);
    }
}

/* -------- Host Code (CPU) -------- */
int main() {

    // (a) Host array
    int h_arr[5] = {10, 20, 30, 40, 50};

    // (b) Allocate device memory
    int *d_arr;
    cudaMalloc(&d_arr, 5 * sizeof(int));

    // (c) Copy data from host to device
    cudaMemcpy(d_arr, h_arr, 5 * sizeof(int), cudaMemcpyHostToDevice);

    // (d) Launch kernel
    printDeviceArray<<<1, 5>>>(d_arr);
    cudaDeviceSynchronize();

    // (e) Copy data back from device to host
    cudaMemcpy(h_arr, d_arr, 5 * sizeof(int), cudaMemcpyDeviceToHost);

    // Print on CPU
    printf("CPU after copy back:\n");
    for (int i = 0; i < 5; i++) {
        printf("CPU: h_arr[%d] = %d\n", i, h_arr[i]);
    }

    // Free device memory
    cudaFree(d_arr);
    cudaDeviceReset();

    return 0;
}
```

```python
!nvcc -arch=sm_75 memory_sep.cu -o memory_sep
```

```python
!./memory_sep
```

Compare CPU times of List/tuple with Numpy arrays.

```python
import numpy as np

N = 1_000_000
lst = list(range(N))
%time lst_result = [x * 2 for x in lst]
```
