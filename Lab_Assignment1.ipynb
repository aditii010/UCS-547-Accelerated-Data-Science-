```python
import numpy as np
x= np.array([1,2,3,4,5,1,2,1,1,1])
frequency = {}
for value in x:
  if value in frequency:
    frequency[value]+=1
  else:
      frequency[value]=1
print (frequency)
most_freq_value = max(frequency, key=frequency.get)
print (most_freq_value)
indices= []
for i in range (len(x)):
  if x[i]==most_freq_value:
     indices.append(i)
print("Most frequent value:", most_freq_value)
print("Indices:", indices)
```

```python
import numpy as np
y=np.array([1,1,1,2,3,4,2,4,3,3])
values, counts = np.unique (y, return_counts=True)
most_frequent_value= values[np.argmax(counts)]
indices = np.where(y == most_freq_value)
print("Most frequent value:", most_freq_value)
print("Indices:", indices[0])

```

argmax: index of largest value


*   return_counts=True : how many times each value appears
*   np.unique: gives all different values in the array

```python
import numpy as np
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print("Given matrix:")
print(gfg)
total_sum = np.sum(gfg)
print("\nSum of all elements:")
print(total_sum)
row_sum = np.sum(gfg, axis=1)
print("\nrow wise sum:")
print(row_sum)
column_sum = np.sum(gfg, axis=0)
print("\ncolumn wise sum:")
print(column_sum)
```
